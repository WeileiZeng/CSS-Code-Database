## Sample entry
There are three files for each entry E
- E.json
  - save parameters
- EGx.mm
  - X type parity-check matrix
- EGz.mm
  - Z type parity-check matrix

````python
#sample read
import scipy
file_prefix='codes/i1/n7/k1/n7k1d3-x3z3dx3dz3-1'
print('Gx:')
Gx=scipy.io.mmread(file_prefix+'Gx.mm')
print(Gx.todense())
print('Gz:')
Gz=scipy.io.mmread(file_prefix+'Gz.mm')
print(Gz.todense())
import json
json.load(open(file_prefix+'.json'))
````

## How to find the code
just follow the filename convention 
`n11k2d3-x4z5dx3dz3-2Gz.mm`,
where the last number is an index among 0-9, which corresponding to 10 instances with same parameters
